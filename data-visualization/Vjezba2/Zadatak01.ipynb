```python
import numpy as np
import matplotlib.pyplot as ptl
```

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linearno')
plt.plot(x, x**2, label = 'kvadratno')
plt.plot(x, x**3, label='kubno')
plt.xlabel('x osa')
plt.ylabel('y osa')
plt.title('Jednostavan crte')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t,t**3, 'g^')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2)
x1 = np.linspace(0, 10 ,100)

ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-')
```
